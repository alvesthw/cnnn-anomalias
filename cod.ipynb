```python
import os
import shutil
import random

# Pasta raiz do dataset atual
origem = r"C:/Users/matbm/OneDrive/Documents/Faculdade-Matheus/TCC/data/images"

# Pasta onde será criado o dataset reorganizado
destino = r"C:/Users/matbm/OneDrive/Documents/Faculdade-Matheus/TCC/data/dataset_reorganizado"

# Classes que você tem
classes = ['Base', 'Code', 'normal', 'Rule(F)', 'Rule(R)']

# Proporção treino/teste
train_ratio = 0.8

for classe in classes:
    imagens = []

    # Buscar imagens em train e test atuais
    for tipo in ['train', 'test']:
        pasta_atual = os.path.join(origem, classe, tipo)
        if os.path.exists(pasta_atual):
            for arquivo in os.listdir(pasta_atual):
                if arquivo.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                    imagens.append(os.path.join(pasta_atual, arquivo))

    # Embaralhar as imagens
    random.shuffle(imagens)

    # Dividir 80/20
    corte = int(len(imagens) * train_ratio)
    train_imgs = imagens[:corte]
    test_imgs = imagens[corte:]

    # Criar pastas destino
    pasta_train = os.path.join(destino, 'train', classe)
    pasta_test = os.path.join(destino, 'test', classe)
    os.makedirs(pasta_train, exist_ok=True)
    os.makedirs(pasta_test, exist_ok=True)

    # Copiar imagens treino
    for img_path in train_imgs:
        shutil.copy2(img_path, pasta_train)

    # Copiar imagens teste
    for img_path in test_imgs:
        shutil.copy2(img_path, pasta_test)

    print(f'Classe {classe}: {len(train_imgs)} treino, {len(test_imgs)} teste')

print('✅ Dataset reorganizado com sucesso!')
```

```python
# Experimento 1

import tensorflow as tf
import matplotlib.pyplot as plt

caminho_dataset = r"C:/Users/matbm/OneDrive/Documents/Faculdade-Matheus/TCC/data/dataset_reorganizado"

# Ajuste o tamanho da imagem aqui para 128x128
tamanho_imagem = (128, 128) # <-- Mudei para 128x128 para otimização de velocidade
batch_size = 32 # Considere aumentar para 64 ou 128 se sua GPU permitir

# Dataset de treino
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    directory=f"{caminho_dataset}/train",
    labels="inferred",
    label_mode="int",
    batch_size=batch_size,
    image_size=tamanho_imagem, # <--- Usando o tamanho_imagem ajustado
    shuffle=True,
    seed=123,
)

# Dataset de teste
test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    directory=f"{caminho_dataset}/test",
    labels="inferred",
    label_mode="int",
    batch_size=batch_size,
    image_size=tamanho_imagem, # <--- Usando o tamanho_imagem ajustado
    shuffle=False,
)

# Exemplo: ver classes mapeadas
print("Classes:", train_dataset.class_names)

# --- OTIMIZAÇÃO DO PIPELINE DE DADOS (CRUCIAL PARA VELOCIDADE) ---
AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.cache().prefetch(buffer_size=AUTOTUNE)


# Data Augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1)
])

# Modelo melhorado
model = tf.keras.Sequential([
    data_augmentation,
    # input_shape agora usa tamanho_imagem[0] e tamanho_imagem[1]
    tf.keras.layers.Rescaling(1./255, input_shape=(tamanho_imagem[0], tamanho_imagem[1], 3)),

    tf.keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(128, 3, activation='relu', padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(256, 3, activation='relu', padding='same'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(5, activation='softmax')  # 5 classes
])

# --- COMPILAR MODELO (BLOCO ADICIONADO/REINSERIDO AQUI) ---
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Redução da LR
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3)
]

# Treinar
history = model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=30,
    callbacks=callbacks
)

# Plotar métricas
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Loss treino')
plt.plot(history.history['val_loss'], label='Loss validação')
plt.title('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Acurácia treino')
plt.plot(history.history['val_accuracy'], label='Acurácia validação')
plt.title('Acurácia')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image

# Caminho para o diretório do dataset
caminho_dataset = Path(r"C:/Users/matbm/OneDrive/Documents/Faculdade-Matheus/TCC/data/dataset_reorganizado")

# Diretórios de Rule(F) e Rule(R) dentro de 'train' ou 'test'
diretorios = [
    caminho_dataset / 'train' / 'Rule(F)',  # Rule(F) do diretório 'train'
    caminho_dataset / 'train' / 'Rule(R)'   # Rule(R) do diretório 'train'
]

# Função para obter uma imagem de cada diretório
def obter_imagem(diretorio, nome_classe):
    # Verifica se o diretório existe
    if not diretorio.exists():
        print(f"Diretório não encontrado: {diretorio}")
        return None
    
    # Lista os arquivos no diretório
    arquivos = [arquivo for arquivo in diretorio.iterdir() if arquivo.suffix in ['.jpg', '.jpeg', '.png', '.bmp']]
    
    if arquivos:
        # Carrega a primeira imagem
        imagem = Image.open(arquivos[0])
        imagem.save(f"{nome_classe}_exemplo.jpg")  # Salva a imagem com um nome identificável
        return imagem
    else:
        print(f"Não foi encontrada nenhuma imagem no diretório: {diretorio}")
        return None

# Obtém uma imagem de cada classe Rule(F) e Rule(R)
imagens = {}
for diretorio in diretorios:
    classe = diretorio.name  # Nome da classe (Rule(F) ou Rule(R))
    imagem = obter_imagem(diretorio, classe)
    if imagem:
        imagens[classe] = imagem

# Exibe as imagens de Rule(F) e Rule(R) usando matplotlib
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # Criando 2 subplots
for i, (classe, imagem) in enumerate(imagens.items()):
    ax[i].imshow(imagem)  # Exibe a imagem
    ax[i].axis('off')  # Desativa os eixos
    ax[i].set_title(f"Exemplo de {classe}")  # Define o título para cada imagem

plt.show()  # Exibe as imagens no gráfico
#
```

```python
#experimento 2

import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, accuracy_score
import numpy as np

# Caminho dos dados
caminho_dataset = r"C:/Users/matbm/OneDrive/Documents/Faculdade-Matheus/TCC/data/dataset_reorganizado"

# Parâmetros
tamanho_imagem = (128, 128)
batch_size = 32
num_classes = 5

# Dataset treino com data augmentation mais agressivo
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal_and_vertical'),
    tf.keras.layers.RandomRotation(0.3),
    tf.keras.layers.RandomZoom(0.3),
    tf.keras.layers.RandomContrast(0.3),
    tf.keras.layers.RandomTranslation(0.2, 0.2),
])

# Carregar datasets
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    directory=f"{caminho_dataset}/train",
    labels="inferred",
    label_mode="int",
    batch_size=batch_size,
    image_size=tamanho_imagem,
    shuffle=True,
    seed=123,
)

test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    directory=f"{caminho_dataset}/test",
    labels="inferred",
    label_mode="int",
    batch_size=batch_size,
    image_size=tamanho_imagem,
    shuffle=False,
)

# Prefetch para otimizar pipeline
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.cache().prefetch(buffer_size=AUTOTUNE)

# Base pré-treinada EfficientNetB0
base_model = tf.keras.applications.EfficientNetB0(
    input_shape=(tamanho_imagem[0], tamanho_imagem[1], 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Congelar inicialmente

# Construindo o modelo
inputs = tf.keras.Input(shape=(tamanho_imagem[0], tamanho_imagem[1], 3))
x = data_augmentation(inputs)
x = tf.keras.layers.Rescaling(1./255)(x)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.5)(x)
outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

# Compilar modelo (fase 1)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-7)
]

# Treinamento inicial (congelado)
history = model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=25,
    callbacks=callbacks
)

# Descongelar parte da base para fine-tuning (exemplo: últimas 30 camadas)
base_model.trainable = True

# Congelar todas as camadas exceto últimas 30
for layer in base_model.layers[:-30]:
    layer.trainable = False

# Compilar para fine-tuning com LR mais baixo
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Fine-tuning
history_fine = model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=15,
    callbacks=callbacks
)

# Avaliação final: métricas e matriz de confusão
y_true = []
y_pred = []

for images, labels in test_dataset:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=train_dataset.class_names)

plt.figure(figsize=(7,7))
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title('Matriz de Confusão - Experimento 2')
plt.show()

print("\nRelatório de Classificação:\n")
print(classification_report(y_true, y_pred, target_names=train_dataset.class_names))

acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')

print(f"Acurácia: {acc:.4f}")
print(f"Precisão (macro): {precision:.4f}")
print(f"Recall (macro): {recall:.4f}")
print(f"F1-Score (macro): {f1:.4f}")
```
